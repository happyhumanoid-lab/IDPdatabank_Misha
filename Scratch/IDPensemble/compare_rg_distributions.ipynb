```python
import numpy as np
import matplotlib.pyplot as plt

def read_rg_values(filename):
    """Read Rg (2nd column) values from a GROMACS gyrate .xvg file."""
    rg_values = []

    with open(filename, 'r') as f:
        for line in f:
            line = line.strip()
            if not line or line.startswith(('#', '@')):
                continue  # Skip header/meta lines
            parts = line.split()
            if len(parts) >= 2:
                try:
                    rg = float(parts[1])
                    rg_values.append(rg)
                except ValueError:
                    continue  # Skip malformed lines

    return np.array(rg_values)

def plot_rg_distribution(rg_values, bins=50):
    """Plot the distribution of Rg values."""
    plt.hist(rg_values, bins=bins, color='steelblue', edgecolor='black', alpha=0.7)
    plt.xlabel("Radius of Gyration (nm)")
    plt.ylabel("Frequency")
    plt.title("Distribution of Rg")
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import jensenshannon
from itertools import combinations
import pandas as pd

def read_rg_values(filename):
    """Read Rg (2nd column) values from a GROMACS gyrate .xvg file."""
    rg_values = []
    with open(filename, 'r') as f:
        for line in f:
            line = line.strip()
            if not line or line.startswith(('#', '@')):
                continue
            parts = line.split()
            if len(parts) >= 2:
                try:
                    rg_values.append(float(parts[1]))
                except ValueError:
                    continue
    return np.array(rg_values)

def plot_rg_distribution(rg_values, label=None, bins=50):
    """Plot a normalized histogram for visualization."""
    plt.hist(rg_values, bins=bins, density=True, alpha=0.5, label=label)

def get_distribution(rg_values, bins):
    """Return normalized histogram (probability distribution)."""
    hist, _ = np.histogram(rg_values, bins=bins, density=True)
    hist = hist / np.sum(hist)
    return hist

# ---------------------------------
# Main analysis
# ---------------------------------
systems_rg = {}
bins = np.linspace(3.0, 4.5, 80)  # Use a consistent bin range for all systems

for system in systems:
    if system['ID'] not in IDs:
        continue

    gyrate_path = databankPath + 'Data/Simulations/' + system['path'] + 'gyrate.xvg'
    rg_values = read_rg_values(gyrate_path)
    label = system['TRJ'][0][0]

    print(label)
    print(f"Read {len(rg_values)} Rg values")
    print(f"Mean Rg: {np.mean(rg_values):.3f} nm, Std: {np.std(rg_values):.3f} nm")

    # Store normalized distribution
    systems_rg[label] = get_distribution(rg_values, bins)
    plot_rg_distribution(rg_values, label=label, bins=bins)

plt.legend()
plt.xlabel("Radius of gyration (nm)")
plt.ylabel("Probability density")
plt.title("Rg Distributions Across Systems")
plt.tight_layout()
plt.show()

# ---------------------------------
# Jensen–Shannon Divergence matrix
# ---------------------------------
labels = list(systems_rg.keys())
n = len(labels)
jsd_matrix = np.zeros((n, n))

for i, j in combinations(range(n), 2):
    p, q = systems_rg[labels[i]], systems_rg[labels[j]]
    jsd = jensenshannon(p, q, base=2)**2
    jsd_matrix[i, j] = jsd
    jsd_matrix[j, i] = jsd  # Symmetric

# Convert to DataFrame for clarity and plotting
jsd_df = pd.DataFrame(jsd_matrix, index=labels, columns=labels)
print("\nJensen–Shannon Divergence Matrix:")
print(jsd_df.round(4))

# ---------------------------------
# Plot heatmap
# ---------------------------------
plt.figure(figsize=(8, 6))
plt.imshow(jsd_df, cmap="viridis", interpolation="nearest")
plt.colorbar(label="Jensen–Shannon Divergence")
plt.xticks(ticks=np.arange(n), labels=labels, rotation=45, ha='right')
plt.yticks(ticks=np.arange(n), labels=labels)
plt.title("Pairwise Jensen–Shannon Divergences Between Systems")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import jensenshannon, squareform
from scipy.cluster.hierarchy import linkage, leaves_list
from itertools import combinations
import pandas as pd

def read_rg_values(filename):
    """Read Rg (2nd column) values from a GROMACS gyrate .xvg file."""
    rg_values = []
    with open(filename, 'r') as f:
        for line in f:
            line = line.strip()
            if not line or line.startswith(('#', '@')):
                continue
            parts = line.split()
            if len(parts) >= 2:
                try:
                    rg_values.append(float(parts[1]))
                except ValueError:
                    continue
    return np.array(rg_values)

def plot_rg_distribution(rg_values, label=None, bins=50):
    """Plot a normalized histogram for visualization."""
    plt.hist(rg_values, bins=bins, density=True, alpha=0.5, label=label)

def get_distribution(rg_values, bins):
    """Return normalized histogram (probability distribution)."""
    hist, _ = np.histogram(rg_values, bins=bins, density=True)
    hist = hist / np.sum(hist)
    return hist

# ---------------------------------
# Main analysis
# ---------------------------------
systems_rg = {}
bins = np.linspace(3.0, 4.5, 80)

for system in systems:
    if system['ID'] not in IDs:
        continue

    gyrate_path = databankPath + 'Data/Simulations/' + system['path'] + 'gyrate.xvg'
    rg_values = read_rg_values(gyrate_path)

    # Convert TRJ to string label
    label = system['TRJ'][0][0]

    print(label)
    print(f"Read {len(rg_values)} Rg values")
    print(f"Mean Rg: {np.mean(rg_values):.3f} nm, Std: {np.std(rg_values):.3f} nm")

    systems_rg[label] = get_distribution(rg_values, bins)
    plot_rg_distribution(rg_values, label=label, bins=bins)

plt.legend()
plt.xlabel("Radius of gyration (nm)")
plt.ylabel("Probability density")
plt.title("Rg Distributions Across Systems")
plt.tight_layout()
plt.show()

# ---------------------------------
# Jensen–Shannon Divergence matrix
# ---------------------------------
labels = list(systems_rg.keys())
n = len(labels)
jsd_matrix = np.zeros((n, n))

for i, j in combinations(range(n), 2):
    p, q = systems_rg[labels[i]], systems_rg[labels[j]]
    jsd = jensenshannon(p, q, base=2)**2
    jsd_matrix[i, j] = jsd
    jsd_matrix[j, i] = jsd

# Convert to DataFrame
jsd_df = pd.DataFrame(jsd_matrix, index=labels, columns=labels)
print("\nJensen–Shannon Divergence Matrix:")
print(jsd_df.round(4))

# ---------------------------------
# Cluster systems by similarity
# ---------------------------------
# Convert to condensed distance matrix for linkage
condensed = squareform(jsd_matrix)
linkage_matrix = linkage(condensed, method='average')
ordered_indices = leaves_list(linkage_matrix)

# Reorder matrix and labels
ordered_labels = [labels[i] for i in ordered_indices]
ordered_jsd = jsd_df.loc[ordered_labels, ordered_labels]

# ---------------------------------
# Plot clustered heatmap
# ---------------------------------
plt.figure(figsize=(8, 6))
plt.imshow(ordered_jsd, cmap="viridis", interpolation="nearest")
plt.colorbar(label="Jensen–Shannon Divergence")
plt.xticks(ticks=np.arange(n), labels=ordered_labels, rotation=45, ha='right')
plt.yticks(ticks=np.arange(n), labels=ordered_labels)
plt.title("Rg Distribution Similarity (Clustered by JSD)")
plt.tight_layout()
plt.show()
```
